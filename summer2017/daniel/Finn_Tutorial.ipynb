# Numpy and Arrays

```python
import numpy
```

```python
data = numpy.loadtxt(fname='data/inflammation-01.csv', delimiter=',')
print(data)
```

```python
print(type(data))
```

This tells us that the data currently refers to an N-dimensional array created by the NumPy library

```python
print(data.dtype)
```

```python
patient_0 = data[0, :] # 0 on the first axis, everything on the second
print('maximum inflammation for patient 0:', patient_0.max())
```

# Plotting

```python
import matplotlib.pyplot
image = matplotlib.pyplot.imshow(data)
matplotlib.pyplot.show()
# If on Ipython notebook command - %matplotlib inline - is needed for plot to 
# appear when show command is given
```

```python
ave_inflammation = numpy.mean(data, axis=0)
ave_plot = matplotlib.pyplot.plot(ave_inflammation)
matplotlib.pyplot.show()
```
