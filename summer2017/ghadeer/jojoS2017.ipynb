```python
def efficiency_plot(logMhalo, alpha, epsilon0):
    fig, ax = plt.subplots(figsize=(10,5))
    for ee in alpha:
        ax.plot(logMhalo,getMstar(logMhalo, epsilon0=epsilon0, alpha=ee), label=r'$\epsilon = {:.1f}$'.format(ee))
    ax.legend(loc='upper left')
    ax.set_xlabel(r'$\log_{10}\ (M_{halo} / M_{\odot})$', fontsize = 18)
    ax.set_ylabel(r'$\log_{10}\ (M_{*} / M_{\odot})$', fontsize = 18)
    #ax.set_xlim(12.5, 15.5)
    #ax.set_ylim(9.5, 15)
    ax.set_title(r'$\epsilon0 = {:0g}$'.format(epsilon0))
    
```

```python
efficiency_plot(logMhalo, alpha, epsilon0=0.5)
```
